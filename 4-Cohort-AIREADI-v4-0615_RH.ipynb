```python
# Process all CGM files from the cgm_json_list and combine into one DataFrame
import os
from tqdm.notebook import tqdm
import pandas as pd
import json
from datetime import datetime

def convert_cgm_json_to_df(json_path):
    """
    Convert CGM JSON data to a pandas DataFrame.
    
    Args:
        json_path (str): Path to the CGM JSON file
        
    Returns:
        pd.DataFrame: DataFrame containing CGM data with all available columns
    """
    
    # Load JSON data
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    # Extract patient ID from header
    patient_id = data['header']['patient_id']
    if 'AIREADI-' not in patient_id:
        patient_id = 'AIREADI-' + patient_id
    
    # Initialize lists to store data
    records = []
    
    # Process CGM readings
    for reading in data['body']['cgm']:
        # Create a base record with patient ID
        record = {'patient_id': patient_id}
        
        # Extract timestamp from effective_time_frame
        if 'effective_time_frame' in reading and 'time_interval' in reading['effective_time_frame']:
            time_interval = reading['effective_time_frame']['time_interval']
            if 'start_date_time' in time_interval:
                record['timestamp'] = time_interval['start_date_time']
        
        # Extract blood glucose value
        if 'blood_glucose' in reading:
            record['glucose_value'] = reading['blood_glucose']['value']
            record['glucose_unit'] = reading['blood_glucose']['unit']
        
        # Add all other available fields
        for key, value in reading.items():
            if key not in ['effective_time_frame', 'blood_glucose']:
                if isinstance(value, dict):
                    for sub_key, sub_value in value.items():
                        record[f"{key}_{sub_key}"] = sub_value
                else:
                    record[key] = value
        
        records.append(record)
    
    # Create DataFrame
    df = pd.DataFrame(records)
    
    # Convert timestamp to datetime if it exists
    if 'timestamp' in df.columns:
        # df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
        df['timestamp'] = pd.to_datetime(df['timestamp'], format='mixed')

    return df

# Test the function with the first CGM JSON file

print(cgm_json_path)
cgm_df = convert_cgm_json_to_df(cgm_json_path)
cgm_df
```

```python
import pandas as pd 

# print(f"Found {len(cgm_json_list)} CGM JSON files")

# Process all CGM files and combine into one DataFrame
all_cgm_data = []

for file_path in tqdm(cgm_json_list, desc="Processing CGM files"):
    try:
        # Extract patient ID from the file path
        patient_id = os.path.basename(os.path.dirname(file_path))
        
        # Convert the JSON file to DataFrame
        df = convert_cgm_json_to_df(file_path)
        
        # Add patient ID if not already in the DataFrame
        if 'patient_id' not in df.columns:
            df['patient_id'] = patient_id
            
        all_cgm_data.append(df)
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

final_cgm_df = pd.concat(all_cgm_data, ignore_index=True)

final_cgm_df.shape 
```

```python
def get_timezone_offset_minutes_corrected(dt, timezone_name):
    """
    Get timezone offset in minutes for a specific datetime
    
    Args:
        dt: datetime object (in UTC, either timezone-aware or naive)
        timezone_name: string like 'America/Los_Angeles'
        
    Returns:
        offset in minutes (negative for timezones west of UTC)
    """
    import pytz
    
    # Ensure dt is timezone-aware UTC
    if dt.tzinfo is None:
        dt = dt.replace(tzinfo=pytz.UTC)
    elif dt.tzinfo != pytz.UTC:
        dt = dt.astimezone(pytz.UTC)
        
    # Get the timezone object
    tz = pytz.timezone(timezone_name)
    
    # Convert to target timezone
    localized_dt = dt.astimezone(tz)
    
    # Get the offset in seconds and convert to minutes
    offset_seconds = localized_dt.utcoffset().total_seconds()
    offset_minutes = int(offset_seconds / 60)
    
    return offset_minutes
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from scipy import stats



# TIME_COL_TO_USE = 'DT_local' # localized time
TIME_COL_TO_USE = 'timestamp'
# Ensure DT_local is datetime
df_all[TIME_COL_TO_USE] = pd.to_datetime(df_all[TIME_COL_TO_USE])

# Function to round time to nearest 5-minute interval
def round_to_5min(dt):
    """Round datetime to nearest 5-minute interval"""
    # Get the minute component
    minute = dt.minute
    # Round to nearest 5-minute mark
    rounded_minute = round(minute / 5) * 5
    # Handle the case where rounding goes to 60 minutes
    if rounded_minute == 60:
        rounded_minute = 0
        dt = dt + timedelta(hours=1)
    
    # Create new datetime with rounded minute, zero seconds/microseconds
    return dt.replace(minute=rounded_minute, second=0, microsecond=0)

# Apply the rounding function
df_all['DT_5min'] = df_all[TIME_COL_TO_USE].apply(round_to_5min)

# Extract just time component for plotting (HH:MM format)
df_all['Time_5min'] = df_all['DT_5min'].dt.strftime('%H:%M')

# Group by MRSegmentID and 5-minute time intervals
grouped_stats = df_all.groupby(['study_group', 'Time_5min'])['glucose_value'].agg([
    'mean',
    'std',
    'count',
    lambda x: np.percentile(x, 5),   # 5th percentile
    lambda x: np.percentile(x, 95)   # 95th percentile
]).reset_index()

# Rename the lambda columns
grouped_stats.columns = ['study_group', 'Time_5min', 'mean', 'std', 'count', 'p5', 'p95']

# Calculate 90% confidence interval using t-distribution
def calculate_90_ci(row):
    if row['count'] > 1:
        # Use t-distribution for small samples
        t_critical = stats.t.ppf(0.95, row['count'] - 1)  # 95th percentile for 90% CI
        margin_error = t_critical * (row['std'] / np.sqrt(row['count']))
        ci_lower = row['mean'] - margin_error
        ci_upper = row['mean'] + margin_error
    else:
        # Single observation - no CI
        ci_lower = row['mean']
        ci_upper = row['mean']
    
    return pd.Series([ci_lower, ci_upper])

# Apply CI calculation
grouped_stats[['ci_lower', 'ci_upper']] = grouped_stats.apply(calculate_90_ci, axis=1)

# Convert Time_5min to datetime for proper plotting
grouped_stats['Time_dt'] = pd.to_datetime(grouped_stats['Time_5min'], format='%H:%M').dt.time

# Sort by time for proper line plotting
grouped_stats = grouped_stats.sort_values(['study_group', 'Time_5min'])
```

```python

# Create the plot
plt.figure(figsize=(12, 8))

# Get unique MRSegmentIDs and assign colors
segment_ids = grouped_stats['study_group'].unique()
colors = plt.cm.tab10(np.linspace(0, 1, len(segment_ids)))

for i, segment_id in enumerate(segment_ids):
    segment_data = grouped_stats[grouped_stats['study_group'] == segment_id].copy()
    
    # Convert time strings to numeric for plotting
    time_numeric = [pd.to_datetime(t, format='%H:%M').hour + pd.to_datetime(t, format='%H:%M').minute/60 
                   for t in segment_data['Time_5min']]
    
    # Plot mean line
    plt.plot(time_numeric, segment_data['mean'], 
             color=colors[i], linewidth=2, label=f'{segment_id}', marker='o', markersize=4)
    
    # Plot 90% confidence interval
    plt.fill_between(time_numeric, 
                     segment_data['ci_lower'], 
                     segment_data['ci_upper'],
                     color=colors[i], alpha=0.2)

# Customize the plot
plt.xlabel('Time of Day (Hours)', fontsize=12)
plt.ylabel('Blood Glucose Value (mg/dL)', fontsize=12)
plt.title('CGM Profiles by MR Segment ID\n(5-minute intervals with 90% Confidence Intervals)', fontsize=14)
plt.grid(True, alpha=0.3)

# Format x-axis to show hours
plt.xticks(range(0, 25, 2), [f'{h:02d}:00' for h in range(0, 25, 2)])
plt.xlim(0, 24)

# Set y-axis limits
plt.ylim(50, 200)

# Add horizontal lines at 70 and 180 mg/dL
plt.axhline(y=70, color='red', linestyle='--', linewidth=1, label='Hypoglycemia Threshold (70 mg/dL)')
plt.axhline(y=180, color='orange', linestyle='--', linewidth=1, label='Hyperglycemia Threshold (180 mg/dL)')

# Place the legend at the bottom of the figure
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=2, fancybox=True, shadow=True)

plt.tight_layout()
plt.show()

# Display summary statistics
print("Summary Statistics by study_group and Time:")
print(grouped_stats.head(10))

# Additional summary: Overall statistics by MRSegmentID
overall_stats = df_all.groupby('study_group')['glucose_value'].agg([
    'count', 'mean', 'std', 'min', 'max'
]).round(2)
print("\nOverall Statistics by study_group:")
print(overall_stats)

# Show the number of unique time points per segment
time_points = grouped_stats.groupby('study_group')['Time_5min'].count()
print(f"\nNumber of 5-minute time points per study_group:")
print(time_points)
```
